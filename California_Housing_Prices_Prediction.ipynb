```python
import pandas as pd

housing_pd = pd.read_csv('/housing.csv')
housing_pd
```

```python
housing_pd_shuffled = housing_pd.sample(n=len(housing_pd), random_state=1)
housing_pd_shuffled
```

```python
pd.get_dummies(housing_pd_shuffled['ocean_proximity']).head()
```

```python
housing_pd_shuffled.drop('ocean_proximity', axis=1).head()
```

```python
housing_pd_shuffled_final = pd.concat([housing_pd_shuffled.drop('ocean_proximity', axis=1), pd.get_dummies(housing_pd_shuffled['ocean_proximity'])], axis=1)
housing_pd_shuffled_final = housing_pd_shuffled_final*1
housing_pd_shuffled_final
```

```python
housing_pd_final = housing_pd_shuffled_final[['longitude',	'latitude',
                                     'housing_median_age',	'total_rooms',
                                     'total_bedrooms',	'population',
                                     'households',	'median_income',
                                     '<1H OCEAN',
                                     'INLAND',	'ISLAND',
                                     'NEAR BAY',	'NEAR OCEAN', 'median_house_value']]
housing_pd_final
```

```python
housing_pd_final=housing_pd_final.dropna()
len(housing_pd_final)
```

```python
train_pd,test_pd,val_pd = housing_pd_final[:15000],housing_pd_final[15000:17500],housing_pd_final[17500:]
len(train_pd),len(test_pd),len(val_pd)
```

```python
X_train,y_train = train_pd.to_numpy()[:,:-1],train_pd.to_numpy()[:,-1]
X_val,y_val = val_pd.to_numpy()[:,:-1],val_pd.to_numpy()[:,-1]
X_test,y_test = test_pd.to_numpy()[:,:-1],test_pd.to_numpy()[:,-1]
```

```python
X_train.shape,y_train.shape,X_val.shape,y_val.shape,X_test.shape,y_test.shape
```

Feature Engineering --- Scaling and Transforming Data

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

scaler = StandardScaler().fit(X_train[:, :8])

def preprocessor(X):
  A = np.copy(X)
  A[:, :8] = scaler.transform(A[:, :8])
  return A

X_train, X_val, X_test = preprocessor(X_train), preprocessor(X_val), preprocessor(X_test)
```

```python
pd.DataFrame(X_train).head()
```

```python
pd.DataFrame(X_train).hist(bins=10, figsize=(20,15))
```

```python
#Training and testing using Linear Regression model
from sklearn.metrics import mean_squared_error as mse
from sklearn.linear_model import LinearRegression

lm = LinearRegression().fit(X_train, y_train)
mse(lm.predict(X_train), y_train), mse(lm.predict(X_val), y_val)
```

```python
#Training and testing using KNN model
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=10).fit(X_train, y_train)
mse(knn.predict(X_train), y_train), mse(knn.predict(X_val), y_val)
```

```python
#Training and testing using Random model
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(max_depth=10).fit(X_train, y_train)
mse(rfr.predict(X_train), y_train), mse(rfr.predict(X_val), y_val)
```

```python
#Training and testing using Gradient Descent model
from sklearn.ensemble import GradientBoostingRegressor

gbr = GradientBoostingRegressor(n_estimators=250).fit(X_train, y_train)
mse(gbr.predict(X_train), y_train), mse(gbr.predict(X_val), y_val)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

simple_nn = Sequential()
simple_nn.add(InputLayer((13,)))
simple_nn.add(Dense(2, activation='relu'))
simple_nn.add(Dense(1, activation='linear'))

simple_nn.compile(optimizer=Adam(learning_rate=0.1), loss='mse', metrics=[RootMeanSquaredError()])
simple_nn.fit(X_train, y_train, epochs=20, validation_data=(X_val, y_val),callbacks=[ModelCheckpoint('models/simple_nn.keras', save_best_only=True)])
```

```python
from tensorflow.keras.models import load_model

simple_nn = load_model('models/simple_nn.keras')
mse(simple_nn.predict(X_train), y_train), mse(simple_nn.predict(X_val), y_val)
```

```python
medium_nn= Sequential()
medium_nn.add(InputLayer((13,)))
medium_nn.add(Dense(32, activation='relu'))
medium_nn.add(Dense(16, activation='relu'))
medium_nn.add(Dense(1, activation='linear'))

medium_nn.compile(optimizer=Adam(learning_rate=0.1), loss='mse', metrics=[RootMeanSquaredError()])
medium_nn.fit(X_train, y_train, epochs=50, validation_data=(X_val, y_val),callbacks=[ModelCheckpoint('models/medium_nn.keras', save_best_only=True)])
```

```python
medium_nn = load_model('models/medium_nn.keras')
mse(medium_nn.predict(X_train), y_train), mse(medium_nn.predict(X_val), y_val)
```
